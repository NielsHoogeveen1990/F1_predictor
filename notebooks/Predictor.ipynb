# 3. Groupby driver and count duels won

```python
def get_duels_won(df):
    return df.assign(
        duels_won = lambda df: df.groupby('driverRef_left')[0]
        .transform(lambda df: df.sum())
    )
```

```python
x = get_duels_won(df)
```

```python
x
```

```python
race_result = x.groupby('driverRef_left')['duels_won'].max().sort_values(ascending=False)
```
